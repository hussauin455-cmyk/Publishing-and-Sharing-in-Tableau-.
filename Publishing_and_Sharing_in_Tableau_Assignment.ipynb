#**Publishing and Sharing in Tableau Assignment**

#1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?

**Tableau Public**:

- **Accessibility**: Highly accessible. Anyone with an internet connection can view and interact with your dashboard without needing an account. It's designed for broad public sharing.
- **Data Security**: Very low. Any data used in your dashboard is publicly visible and downloadable. You should never publish sensitive or proprietary data to Tableau Public. There are no row-level security or robust authentication options.   

**Tableau Server/Cloud:**

- **Accessibility**: Controlled accessibility. Access is restricted to licensed users within your organization or designated external users. Viewers need to log in with credentials.    

- **Data Security**: High. Offers robust security features, including user authentication, permissions, row-level security, and data encryption. You have fine-grained control over who can see what data and what actions they can perform (view, download, edit, etc.). It's suitable for sensitive and proprietary data.

# 2. **When publishing a Tableau dashboard, what happens to a live connection compared to an extract?How does this affect data refresh and performance?**



**Live Connection**:

- **What happens**: When you publish a dashboard with a live connection, Tableau maintains a direct link to the underlying data source (e.g., a database, data warehouse). Tableau Server/Cloud will query this data source directly whenever the dashboard is opened or interacted with.  

- **Data Refresh**: Data is always up-to-date. Every time a user interacts with the dashboard, Tableau fetches the latest data from the source. There's no separate 'refresh' schedule needed for the data itself, as it's always live.   

- **Performance**: Can be slower, especially with large datasets or complex queries, as Tableau is constantly querying the source. Performance heavily depends on the speed and capacity of the underlying data source and the network connection to it.  

**Extract**:

- **What happens**: When you publish a dashboard using an extract, Tableau creates a static snapshot (a highly optimized compressed file) of the data from the original source and stores it within Tableau Server/Cloud. The dashboard then queries this stored extract, not the original data source.   

- **Data Refresh**: Data is not live. The dashboard will show data as it was at the time the extract was created or last refreshed. To get fresh data, the extract itself needs to be refreshed on a schedule (e.g., daily, hourly). Tableau Server/Cloud can be configured to automatically refresh extracts.   

- **Performance**: Generally much faster than live connections. Since the data is pre-processed, optimized, and stored locally on Tableau Server/Cloud, dashboards can load and respond more quickly. Performance is independent of the original data source's performance once the extract is created.

# **3.In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users can view, interact with, or edit a published dashboard?**

- In Tableau Server or Cloud, the feature that allows administrators to control which users can view, interact with, or edit a published dashboard is **Permissions**.

Administrators can set permissions at various levels (site, project, workbook, data source, and individual view/dashboard) to define:

- **Who** can access the content (specific users, groups, or roles).    

- **What** they can do with it (e.g., view, interact, download, edit, comment, save, move, delete).

# **4.If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?**

- If you want users to view a dashboard but not download the underlying data in Tableau Server, you would adjust the **Download Summary Data** and/or **Download Full Data** permissions. You should:

- Deny the Download Summary Data capability.
- Deny the Download Full Data capability.

# **6.List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.**

- When preparing a dashboard for public sharing (especially on platforms like Tableau Public), data security and privacy are paramount. Here are two best practices you should follow:

- 1.**Never Share Sensitive or Personally Identifiable Information (PII)**: This is the most crucial rule. Before publishing, thoroughly review all data in your dashboard to ensure it does not contain any sensitive information, such as names, addresses, financial details, health records, or other PII. If your analysis requires such data, you must aggregate it, anonymize it, or exclude it entirely before publishing publicly. Assume anything you publish will be visible to everyone and could be downloaded.    


- 2.**Aggregate Data to a Sufficient Level**: Instead of sharing raw, detailed transactional data, aggregate it to a higher level (e.g., daily totals instead of individual sales, regional averages instead of specific store locations, demographic groups instead of individual people). Aggregation helps to obscure individual data points, making it much harder to identify specific entities or infer private information, while still allowing for meaningful analysis and visualization.

# **7.After publishing a dashboard, users report a “Data not found” error.What is the most common cause related to data source settings?**

Here's why and what it means:

- **Missing Credentials**: If the data source requires authentication (e.g., a database username and password) and these credentials were not embedded or correctly configured when the workbook was published, Tableau Server/Cloud will not be able to connect to the underlying data source.  

- **Incorrect Credentials**: Even if credentials were provided, they might be outdated, expired, or simply incorrect, leading to a failed connection.

- **Insufficient Permissions for the Service Account**: In some enterprise setups, Tableau Server might use a dedicated service account to access data sources. If this service account does not have the necessary permissions on the database or data warehouse, the connection will fail.  

**Other related causes could include8**:

- **Network connectivity issues**: Tableau Server/Cloud cannot reach the data source host.  

- **Data source moved or renamed**: The path or name of the data source has changed, making the connection string invalid.  

- **Database server is down or unresponsive**.   
  
  **To resolve this, administrators typically need to**:

1. **Go to the published data source or workbook on Tableau Server/Cloud**.  

2. **Edit the connection and re-enter/update the embedded credentials.**   

3. **Ensure the credentials used have sufficient permissions to access the data**.

# **8.When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded content?**

When embedding a Tableau dashboard into a secure internal portal, beyond the default security of Tableau Server/Cloud and the portal itself, two crucial additional security measures to implement are:

- **1.SAML (Security Assertion Markup Language) or OpenID Connect (OIDC) Integration**: This ensures that the authentication process for accessing the embedded dashboard is tightly integrated with the internal portal's existing identity provider. Instead of users needing separate Tableau credentials, they are authenticated via the portal's single sign-on (SSO) system. This centralizes user management, enforces consistent security policies, and simplifies the user experience by eliminating multiple logins.

- 2.**Row-Level Security (RLS) within Tableau**: Even if users are authenticated to view the dashboard, you must ensure they only see data relevant and authorized for them. RLS, implemented either within Tableau (e.g., using user filters or entitlements in the data source) or at the database level, dynamically filters the data presented based on the logged-in user's identity. This prevents users from seeing sensitive data points they are not cleared to access, even if they have permission to view the overall dashboard.
